<b>Import all essential packages

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

<b>Get the the html of the wikipedia page

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'lxml')
df=pd.DataFrame(columns=['PostalCode','Borough','Neighbourhood'])
df
```

<b>Scrape the wikipedia table

```python
for data in soup.find_all('tr'):
    row=data.text
    newrow=row.split('\n')
    if(len(newrow)!=5):break
    if(newrow[2]=='Not assigned'or newrow[3]=='Not assigned'):
        continue
    df=df.append({'PostalCode': newrow[1], 'Borough': newrow[2],'Neighbourhood': newrow[3]},ignore_index=True)
df
```

<b>Remove unecessary rows

```python
df=df.drop([0])
df
```

```python
df.reset_index(inplace=True)
```

```python
df=df.drop('index',axis=1)
```

```python
df['Neighbourhood']=df['Neighbourhood'].str.cat(sep=',')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df=df.drop('index',axis=1)
```
